**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.linspace(1,20,10)
y=np.linspace(2,20,10)
df=pd.DataFrame ({
    'Data 1':x,
    'Data 2':y
})
df.head()
```

```python
#lineplot

sns.lineplot(x='Data 1',y='Data 2', data=df)
plt.show()
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.lineplot(x='day',y='total_bill',hue="sex",data=tips,style ="time")
plt.show
```

**Bar plot**

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
sns.barplot(x='sex',y='fare',data=titanic, color='red',hue='who')
plt.show()
```

```python
#facecolor and edge Color
sns.barplot(x='sex',y='fare',data=titanic, color='red',facecolor='yellow', edgecolor="0.2")
plt.show()
```

```python
#Histogram
tips.head()
```

```python
sns.histplot(x='total_bill',data=tips, discrete='false',kde=True)
plt.show()
```

```python
#Scatter Plot
exs=sns.load_dataset('exercise')
exs.head()
```

```python
sns.scatterplot(x='time',y='pulse',data=exs,hue='kind', alpha=0.5, marker ='s',palette='GnBu')  #alpha for transparency
plt.show()
```

```python
#HeatMap
data=np.random.randint(1,100, size=(10,10))
print(data)
```

```python
sns.heatmap(data)
plt.show()
```

```python
sns.heatmap(data, annot=True, cmap='YlGnBu', alpha=1) #annot - annotation , its shows the number
plt.show()
```

```python
sns.heatmap(data=data,xticklabels=False,yticklabels=False,cbar=False,annot=True)
plt.show()
```

```python
#Countplot
sns.countplot(x='sex',data=tips)
plt.show()
```

```python
sns.countplot(y='sex',data=tips)
plt.show()
```

```python
sns.countplot(x='sex',data=tips,hue ='smoker', palette="GnBu")
plt.show()
```

```python
#Violin Plot
sns.violinplot(x='sex',y='total_bill',data=tips, hue='day')
plt.show()
```

**BoxPlot**

```python
#BoxPlot
sns.boxplot(x='tip',data=tips)
plt.show()
```

```python
sns.boxplot(x='sex',y='tip',data=tips,hue='smoker')
plt.show()
```

```python
#pairplot
sns.pairplot(tips)
plt.show()
```

```python
#cat plot -categorical plot
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm')
plt.show()
```

```python
#striplot
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
plt.show()
```

```python
#KDE plot -kernal density estimation
x=np.random.randn(100)
y=np.random.randn(100)
sns.kdeplot(x)
sns.kdeplot(y,shade=True,color='Red')
plt.show()
```

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
x = np.random.randn(100)
y = np.random.randn(100)

# Create a 2D KDE plot using sns.jointplot()
sns.jointplot(x=x, y=y, kind='kde', shade=True, shade_lowest=False)
plt.show()
```
