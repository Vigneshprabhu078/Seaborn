**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.linspace(1,20,10)
y=np.linspace(2,20,10)
df=pd.DataFrame ({
    'Data 1':x,
    'Data 2':y
})
df.head()
```

```python
#lineplot

sns.lineplot(x='Data 1',y='Data 2', data=df)
plt.show()
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.lineplot(x='day',y='total_bill',hue="sex",data=tips,style ="time")
plt.show
```

**Bar plot**

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
sns.barplot(x='sex',y='fare',data=titanic, color='red',hue='who')
plt.show()
```

```python
#facecolor and edge Color
sns.barplot(x='sex',y='fare',data=titanic, color='red',facecolor='yellow', edgecolor="0.2")
plt.show()
```

```python
#Histogram
tips.head()
```

```python
sns.histplot(x='total_bill',data=tips, discrete='false',kde=True)
plt.show()
```
